```python
def adding_prefix(df_g):
    df_g.replace('http://www.project-improvit.de/Donor/', 'Donor:', regex=True, inplace=True)
    df_g.replace('http://www.project-improvit.de/vocab/', 'improvit:', regex=True, inplace=True)
    #     df_g.replace('http://www.w3.org/1999/02/22-rdf-syntax-ns#', 'type:', regex=True, inplace=True)
    df_g.replace('<http://www.project-improvit.de/Measurement/', 'Measurement:', regex=True, inplace=True)
    df_g.replace('http://www.project-improvit.de/FunctionalCure/', '', regex=True, inplace=True)
    df_g.replace('http://www.project-improvit.de/Parameter/', 'Parameter:', regex=True, inplace=True)
    df_g.replace('http://www.w3.org/2001/XMLSchema#float', '', regex=True, inplace=True)
    df_g.replace('http://www.w3.org/2001/XMLSchema#string', '', regex=True, inplace=True)
    df_g.replace('http://www.w3.org/2001/XMLSchema#int', '', regex=True, inplace=True)
    df_g.replace('http://www.w3.org/1999/02/22-rdf-syntax-ns#', 'rdf:', regex=True, inplace=True)    
    df_g.replace('^^', '', regex=True, inplace=True)
    df_g.replace('>', '', regex=True, inplace=True)
    df_g.replace('<', '', regex=True, inplace=True)
    return df_g
```

```python
kg = pd.read_csv('All_Measurement_parameters_D0_HBsRE_v5.nt', delimiter=" ", header=None)
kg.columns = ['s', 'p', 'o']
kg['o'] = kg['o'].map(lambda x: str(x)[:-1])
kg = adding_prefix(kg)
display(kg.shape, kg.head())
kg.to_csv('network_08.tsv', index=None, header=None, sep='\t')
```
